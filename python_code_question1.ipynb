```python
from dash import Dash, dcc, html
from dash.dependencies import Input, Output
from dash.exceptions import PreventUpdate
import plotly.express as px
import pandas as pd

app = Dash(__name__)


# Regrouper les donnÃ©es pour obtenir le compte d'Ã©vÃ©nements
event_counts = sudan_conflict_data.groupby('event_type').size().reset_index(name='count')
event_counts = event_counts.sort_values(by='count', ascending=False)


# CrÃ©er le graphique principal des types d'Ã©vÃ©nements
fig_event = px.bar(event_counts, 
                   x='event_type', 
                   y='count', 
                   color='event_type',
                   title='Number of conflicts by the type of events',
                   labels={'count': 'Number of conflicts', 'event_type': 'Event Type'},
                   template='plotly_white')

# Disposition de l'application
app.layout = html.Div([
    dcc.Graph(id='event-type-bar', figure=fig_event),
    dcc.Graph(id='sub-event-type-bar'),
])

@app.callback(
    Output('sub-event-type-bar', 'figure'),
    Input('event-type-bar', 'clickData')
)
def display_sub_event_graph(clickData):
    if clickData is None:
        raise PreventUpdate

    selected_event_type = clickData['points'][0]['x']
    filtered_data = sudan_conflict_data[sudan_conflict_data['event_type'] == selected_event_type]
    
    # Regrouper pour obtenir le comptage des sous-types d'Ã©vÃ©nements
    sub_event_counts = filtered_data.groupby('sub_event_type').size().reset_index(name='count')

    # CrÃ©er le graphique pour les sous-types d'Ã©vÃ©nements
    fig_sub_event = px.bar(sub_event_counts, 
                           x='sub_event_type', 
                           y='count', 
                           color='sub_event_type',
                           title=f'Counts of Sub-Event Types for {selected_event_type}',
                           labels={'count': 'Number of Events', 'sub_event_type': 'Sub-Event Type'},
                           template='plotly_white')
    
    return fig_sub_event

if __name__ == '__main__':
    app.run_server(debug=True)
```

```python
import pandas as pd 
import matplotlib.pyplot as plt

# Sample data loading (replace this with your actual data)
# sudan_conflict_data = pd.read_csv('your_data.csv')

# Ensure 'event_date' is in datetime format 
sudan_conflict_data['event_date'] = pd.to_datetime(sudan_conflict_data['event_date'], format='%d %B %Y')

# Get the top 5 locations with the most conflict events
top_locations = (
    sudan_conflict_data['location'].value_counts()
    .head(5)
    .index
)

# Filter the original dataset for these top locations
top_locations_data = sudan_conflict_data[sudan_conflict_data['location'].isin(top_locations)]

# Group by month-year and location, counting occurrences
location_trends = (
    top_locations_data
    .groupby([top_locations_data['event_date'].dt.to_period('M'), 'location'])
    .size()
    .reset_index(name='count')
)

# Convert period back to datetime for plotting
location_trends['event_date'] = location_trends['event_date'].dt.to_timestamp()

# Create a bar plot
plt.figure(figsize=(12, 6))
for location in top_locations:
    loc_data = location_trends[location_trends['location'] == location]
    plt.bar(loc_data['event_date'].astype(str), loc_data['count'], label=location)

plt.title('Trends in Conflict Hotspots: Top 5 Most Affected Locations', fontsize=14)
plt.xlabel('Month-Year', fontsize=12)
plt.ylabel('Number of Events', fontsize=12)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.legend(title='Location')
plt.grid(axis='y')
plt.tight_layout()  # Adjust layout to prevent clipping of labels

# Save the figure as a static image (e.g., PNG)
plt.savefig('evolution_conflict_events.png')

# Show the static plot
plt.show()
```
