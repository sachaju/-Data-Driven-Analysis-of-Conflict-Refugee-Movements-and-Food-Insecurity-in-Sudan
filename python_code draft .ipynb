```python
import geopandas as gpd
import matplotlib.pyplot as plt
# Charger le fichier shapefile
sudan = gpd.read_file("C:/Users/sacha/Downloads/SD_202403/SD_202403_ML2.shp")

# Dissoudre les gomtries en fonction de 'ADMIN1' pour avoir un polygone par rgion
sudan_dissolved = sudan.dissolve(by='ADMIN2')

# Crer la carte
fig, ax = plt.subplots(figsize=(10, 10))
sudan_dissolved.plot(column='ML2', cmap='OrRd', legend=True, ax=ax)

# Ajouter les noms des rgions (ADMIN1) une seule fois au centre de chaque rgion
for idx, row in sudan_dissolved.iterrows():
    # Calculer le centroid de chaque rgion
    centroid = row['geometry'].centroid
    # Annoter le nom de la rgion (ADMIN1) au niveau du centroid
    plt.annotate(text=row.name, xy=(centroid.x, centroid.y),
                 horizontalalignment='center', fontsize=8, color='black', weight='bold')

# Afficher la carte
plt.title('Sudan Food Insecurity Classification by Region (Jun 2024 - Sept 2024)')
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load the shapefile
sudan = gpd.read_file("C:/Users/sacha/Downloads/SD_202403/SD_202403_ML1.shp")

# Dissolve geometries by 'ADMIN2' to have one polygon per region
sudan_dissolved = sudan.dissolve(by='ADMIN2')

# Specify the regions you want to label
regions_to_label = ['Khartoum', 'Omdurman', 'Bahri', 'Al Fasher', 
                    'Karrari', 'Jebel Awlia', 'Sharg An Neel', 
                    'Al Hasahisa', 'Um Bada', 'Nyala Janoub']

# Create the map
fig, ax = plt.subplots(figsize=(10, 10))
sudan_dissolved.plot(column='ML1', cmap='OrRd', legend=True, ax=ax)

# Add the names of specified regions at their centroids
for idx, row in sudan_dissolved.iterrows():
    # Check if the region is in the specified list
    if row.name in regions_to_label:
        # Calculate the centroid of each region
        centroid = row['geometry'].centroid
        # Annotate the name of the region at the centroid
        plt.annotate(text=row.name, xy=(centroid.x, centroid.y),
                     horizontalalignment='center', fontsize=8, color='black', weight='bold')

# Show the map
plt.title('Sudan Food Insecurity Classification by Region (Jun 2024 - Sept 2024)')
plt.show()
```
