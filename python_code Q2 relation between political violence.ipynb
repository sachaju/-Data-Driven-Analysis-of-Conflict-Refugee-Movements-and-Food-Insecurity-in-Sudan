```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates

# Example DataFrame creation (replace with your actual DataFrame)
# merged_data should already contain the columns 'event_date', 'disorder_type', and 'TOTAL'

# Convert 'event_date' to datetime
merged_data['event_date'] = pd.to_datetime(merged_data['event_date'])

# Filter for "Political violence"
pol_violence = merged_data[merged_data['disorder_type'] == 'Political violence']

# Count political violence incidents by month
monthly_political_violence = pol_violence.groupby(pol_violence['event_date'].dt.to_period("M")).sum().reset_index(name='event_count')

# Sum total refugees by month
monthly_refugees = merged_data.groupby(merged_data['event_date'].dt.to_period("M"))['TOTAL'].size().reset_index(name='refugees_count')

# Convert periods to timestamps for plotting
monthly_political_violence['event_date'] = monthly_political_violence['event_date'].dt.to_timestamp()
monthly_refugees['event_date'] = monthly_refugees['event_date'].dt.to_timestamp()

# Merge both DataFrames on event_date
merged_monthly_data = pd.merge(monthly_political_violence, monthly_refugees, on='event_date')

# Calculate correlation
correlation = merged_monthly_data[['event_count', 'refugees_count']].corr()

# Create heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f', cbar=True, square=True, 
            xticklabels=['Political Violence', 'Total Refugees'], 
            yticklabels=['Political Violence', 'Total Refugees'])
plt.title('Heatmap of Correlation between Political Violence and Refugees (2023-2024)', fontsize=16)
plt.show()
```
